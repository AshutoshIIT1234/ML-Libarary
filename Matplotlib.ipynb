Matplotlib


```python
#importing matplotlib libarary
import matplotlib.pyplot as plt
```

```python
#import numpy to get data
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y=np.sin(x)
z=np.cos(x)
```

```python
print (x)
```

```python
print(z)
```

Plotting the Data

```python
plt.plot(x,y)
plt.show
```

```python
#cosine wave
plt.plot(x,z)
plt.show
```

```python
#adding title x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin wave')
plt.show
```

```python
#parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show
```

```python
plt.plot(x,y,'r*')
plt.show()
```

```python
plt.plot(x,y,'gx')
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g+')
plt.plot(x,np.cos(x),'r*')
plt.legend()
plt.show()
```

Bar plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
langs=['c','c++','java','python','php']
students=[23,17,35,29,12]
ax.bar(langs,students)
plt.show()
```

Pie Chart


```python
fig1=plt.figure()
ax1=fig1.add_axes([0,0,1,1])
langs=['c','c++','java','python','php']
students=[23,17,35,29,12]
ax1.pie(students,labels=langs,autopct='%1.2f%%')
plt.show()
```

Scatter Plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax2=fig2.add_axes([0,0,1,1])
ax2.scatter(x,y)
ax2.scatter(x,z)
plt.show()
```

3d scatter Plot

```python
fig3=plt.figure()
ax3=plt.axes(projection='3d')
z= 20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax3.scatter(x,y,z,c=z,cmap='Greens')
plt.show()
```
