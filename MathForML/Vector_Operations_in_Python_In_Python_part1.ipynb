```python
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
sns.set()
```

Plotting a vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,angles='xy',scale_units='xy',scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,angles='xy',scale_units='xy',scale=1, color='b')
plt.quiver(0,0,8,-3,angles='xy',scale_units='xy',scale=1, color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
sum=vector_1+vector_2
print(sum)

```

```python
plt.quiver(0,0,2,3,angles='xy',scale_units='xy',scale=1, color='b')
plt.quiver(0,0,3,-2,angles='xy',scale_units='xy',scale=1, color='y')
plt.quiver(0,0,5,1,angles='xy',scale_units='xy',scale=1, color='')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Multiplying a vector by a scalar

```python
vector_1=np.asarray([0,0,2,3])
scalar=2
product=vector_1*scalar
print(product)
```

```python
plt.quiver(0,0,2,3,angles='xy',scale_units='xy',scale=1, color='b')
plt.quiver(0,0,4,6,angles='xy',scale_units='xy',scale=1, color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1=np.asarray([0,0,2,3])
scalar=-0.5
product=vector_1*scalar
print(product)
```

```python
plt.quiver(0,0,2,3,angles='xy',scale_units='xy',scale=1, color='y')
plt.quiver(0,0,-1,-1.5,angles='xy',scale_units='xy',scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
