```python
#importing the libararies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Note:Seaborn has some built-in datasets

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#visulize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
#load the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

Scatter Plot

```python
sns.scatterplot(data=iris,x='sepal_length',y='petal_length',hue='species',style='species')
```

```python
sns.scatterplot(data=iris,x='sepal_length',y='petal_width',hue='species',style='species')
```

```python
#loading the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

Count Plot

```python
sns.countplot(data=titanic,x='class')
```

```python
sns.countplot(data=titanic,x='survived')
```

Bar Chart


```python
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
#house price dataset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
house = pd.DataFrame(housing.data,columns=housing.feature_names)
house['Price'] = housing.target
house.head()
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
df = pd.DataFrame(boston.data,columns=boston.feature_names)
df['target'] = boston.target
df.head()
```

Distribution Plot

```python
sns.displot(house['Price'])
```
